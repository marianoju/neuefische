# Old Faithful Geyser Dataset 

For a detailed description please see https://stat.ethz.ch/R-manual/R-devel/library/datasets/html/faithful.html.

The Old Faithful Geyser Dataset contains 272 observations on 2 variables.

- eruptions numeric Eruption time in mins
- waiting numeric Waiting time to next eruption (in mins)
---

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
data = pd.read_json("data/old-faithful-geyser/index.json") 
```

```python
data.head()
```

```python
plt.figure(figsize=(20, 10))
plt.scatter(data.w, data.d) 
# see https://stackoverflow.com/questions/14300137/
```

```python
kmeans = KMeans(n_clusters=2, random_state=0).fit(data)
data['cluster_nr'] = kmeans.predict(data)

# Zeichnen:
plt.figure(figsize=(20, 10))
plt.scatter(data.w, data.d, c=data['cluster_nr'])
plt.xlabel('d')
plt.ylabel('w')
plt.title('Old Faithful')
plt.show()
```
