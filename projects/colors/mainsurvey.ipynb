```python
import sqlite3 as sql
import pandas as pd
import re
```

```python
connection = sql.connect('mainsurvey.db')
```

```python
# read SQL database into Pandas DataFrame 
df = pd.read_sql('SELECT * FROM answers WHERE colorname LIKE "%blue%" LIMIT 10;', connection)
```

```python
df.head(10)
```

```python
# select columns from SQL database 
pd.read_sql('SELECT r, g, b, colorname FROM answers LIMIT 5;', connection)
```

```python
# select all rows with a matching value and count occurrences 
df = pd.read_sql('SELECT colorname, COUNT(*) AS occurrences ' +
                 'FROM answers WHERE colorname LIKE "%blue%" ' +
                 'GROUP BY colorname;', connection)
```

```python
df.head(10)
```

```python
# count occurrences of colornames in the table 'answers'
# with a partial match for 'blue' 
# and give as output all groups with more than 3 occurrences 
# sorted in descending order 
pd.read_sql_query(
    'SELECT ' +
        'colorname, ' +
        'COUNT(*) AS occurrences ' +
    'FROM answers ' +
    'WHERE colorname LIKE "%blue%" ' +
    'GROUP BY colorname ' +
    'HAVING occurrences > 3 ' +
    'ORDER BY occurrences DESC; '
    , connection)
```

```python
# https://stackoverflow.com/questions/5071601/
def match(expr, item):
    return re.match(expr, item) is not None
conn = sql.connect('mainsurvey.db')
conn.create_function("MATCHES", 2, match)
cursor = conn.cursor()
cursor.execute("SELECT MATCHES('busy', 'busy or not, here I come');")
print(cursor.fetchone()[0])
cursor.close()
conn.close()
```
