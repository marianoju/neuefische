```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
np.random.seed(seed=1)
```

```python
x1 = np.random.normal(-4, 0.5, 500)
y1 = np.random.normal(-4, 0.5, 500)
x2 = np.random.normal(-4, 0.5, 500)
y2 = np.random.normal(+4, 0.5, 500)
x3 = np.random.normal(-3, 0.5, 500)
y3 = np.random.normal(-1, 0.5, 500)
x4 = np.random.normal(-1, 0.5, 500)
y4 = np.random.normal(-3, 0.5, 500)
x5 = np.random.normal(-1, 0.5, 500)
y5 = np.random.normal(+4, 0.5, 500)
x6 = np.random.normal(+3, 1.5, 500)
y6 = np.random.normal(-4, 1.5, 500)
x7 = np.random.normal(+3, 0.5, 500)
y7 = np.random.normal(+3, 0.5, 500)

x = np.concatenate([x1, x2, x3, x4, x5, x6, x7])
y = np.concatenate([y1, y2, y3, y4, y5, y6, y7])
```

```python
df = pd.DataFrame()

df['x'] = x
df['y'] = y
```

```python
plt.figure(figsize=(15, 15))
plt.scatter(x1, y1) 
plt.scatter(x2, y2) 
plt.scatter(x3, y3) 
plt.scatter(x4, y4) 
plt.scatter(x5, y5) 
plt.scatter(x6, y6) 
plt.scatter(x7, y7) 
```

```python
df.to_csv('data/out.csv', sep=',', index=False)
```

```python
# Computing clusters 
kmeans = KMeans(n_clusters=7, random_state=0).fit(df)
df['cluster_nr'] = kmeans.predict(df)

# Plotting data with colorized clusters
plt.figure(1, figsize=(15, 15))
plt.scatter(x, y, c=df['cluster_nr'])
plt.show();
```
