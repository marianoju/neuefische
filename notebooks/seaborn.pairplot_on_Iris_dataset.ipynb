# seaborn.pairplot on Iris dataset 

see https://seaborn.pydata.org/generated/seaborn.pairplot.html

```python
import seaborn as sns; sns.set(style="ticks", color_codes=True)
```

```python
iris = sns.load_dataset("iris")
```

```python
iris.head()
```

```python
g = sns.pairplot(iris, hue="species", markers="+", \
                 hue_order=["virginica", "versicolor", "setosa"], palette="husl")
```
