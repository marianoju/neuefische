```python
import numpy as np
import matplotlib.pyplot as plt
import scipy, scipy.stats
```

```python
# Throwing two dices and adding the result every time 
dice1 = np.random.randint(low=1, high=7, size=(1000))
dice2 = np.random.randint(low=1, high=7, size=(1000))
sum_of_dices = dice1+dice2
```

```python
# Display the results in a histogram 
# see https://matplotlib.org/api/_as_gen/matplotlib.pyplot.hist.html
plt.figure(figsize=(10, 5))
plt.hist(sum_of_dices, 11, density=True)
plt.show()
```

```python
# Generate evenly spaced values within a given interval. 
# see https://docs.scipy.org/doc/numpy/reference/generated/numpy.arange.html
foo = np.arange(-10, 10, 0.1, dtype=float)
```

```python
plt.figure(figsize=(10, 1))
plt.hist(foo, 100, density=True)
plt.show()
```

```python
# Compute the probability density function for the range of all values 
pdf = scipy.stats.norm.pdf(foo,0,1)
```

```python
plt.plot(foo, pdf)
```

```python
s = np.random.uniform(-10,10,1000)
```

```python
plt.figure(figsize=(10, 5))
plt.hist(s, 100, density=True)
plt.show()
```
