```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
x1 = np.random.normal(1.1, 2, 500)
y1 = np.random.normal(1.1, 2, 500)
x2 = np.random.normal(-1.-1, 1, 500)
y2 = np.random.normal(-1.-1, 1, 500)
x = np.concatenate([x1, x2])
y = np.concatenate([y1, y2])
```

```python
df = pd.DataFrame()

df['x'] = x
df['y'] = y
```

```python
df.describe()
```

```python
plt.figure(figsize=(15, 15))
plt.scatter(x1, y1) 
plt.scatter(x2, y2) 
```

```python
kmeans = KMeans(n_clusters=3, random_state=0).fit(df)
df['cluster_nr'] = kmeans.predict(df)

# plotte ich diesen Datensatz mit 'scatter'
plt.figure(1, figsize=(15, 15))
plt.scatter(x, y, c=df['cluster_nr'])
plt.show();
```
