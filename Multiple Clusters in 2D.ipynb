# Multiple Clusters in 2D 

For an excercise I will create a dataset with 7 clusters with x and y values. 
This dataset will then be exported (for a third pary to analyse). 
K-Means clustering will then be used to compute the clusters. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# In order to generate replicable pseudo-random data a seed is defined 
# see https://docs.scipy.org/doc/numpy/reference/generated/numpy.random.seed.html

np.random.seed(seed=1)
```

```python
# 7 * 2 Numpy Arrays corresponding to 7 times x and y values 

x1 = np.random.normal(-4, 0.25, 500)
y1 = np.random.normal(-4, 0.25, 500)
x2 = np.random.normal(-4, 0.5, 500)
y2 = np.random.normal(+4, 1.5, 500)
x3 = np.random.normal(-3, 0.5, 500)
y3 = np.random.normal(-1, 0.5, 500)
x4 = np.random.normal(-1, 0.5, 500)
y4 = np.random.normal(-3, 0.5, 500)
x5 = np.random.normal(-1, 0.5, 500)
y5 = np.random.normal(+4, 0.5, 500)
x6 = np.random.normal(+3, 1.5, 500)
y6 = np.random.normal(-4, 1.5, 500)
x7 = np.random.normal(+3, 1.5, 500)
y7 = np.random.normal(+3, 0.5, 500)

# All x and y values are concatenated respectively 

x = np.concatenate([x1, x2, x3, x4, x5, x6, x7])
y = np.concatenate([y1, y2, y3, y4, y5, y6, y7])
```

```python
# x and y values are used to create a DataFrame 

df = pd.DataFrame()

df['x'] = x
df['y'] = y
```

```python
# Each cluster is plotted seperately to visualize the distribution and overlap of clusters 
# for color names see https://matplotlib.org/examples/color/named_colors.html

plt.figure(figsize=(15, 15))
plt.scatter(x1, y1, c='m') 
plt.scatter(x2, y2, c='k') 
plt.scatter(x3, y3, c='g') 
plt.scatter(x4, y4, c='c') 
plt.scatter(x5, y5, c='y') 
plt.scatter(x6, y6, c='r') 
plt.scatter(x7, y7, c='b') 
```

```python
# df.to_csv('data/out.csv', sep=',', index=False)
```

```python
# Computing clusters 
kmeans = KMeans(n_clusters=7, random_state=0).fit(df)
df['cluster_nr'] = kmeans.predict(df)

# Plotting data with colorized clusters
plt.figure(1, figsize=(15, 15))
plt.scatter(x, y, c=df['cluster_nr'])
plt.show();
```

As can be seen by comparing the two plots, k-means clustering has difficulties differentiating overlapping clusters, especially with different variance.
